```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cost function
#f(x) = x^2 + x + 1
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3,stop=3,num=100)
x_1
```

```python
# derative of f(x) to find slope 
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[15,5])

# 1st chart cost function

plt.subplot(1,2,1)
plt.title("cost function")
plt.plot(x_1,f(x_1),color="blue",linewidth=3)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel("x",fontsize=16)
plt.ylabel("f(x)",fontsize=16)

#2nd chart for derative

plt.subplot(1,2,2)
plt.title("slope of cost function")
plt.xlim([-2,3])
plt.ylim(-3,6)
plt.plot(x_1,df(x_1),color="skyblue",linewidth=3)
plt.xlabel("x",fontsize=16)
plt.ylabel("d(x)",fontsize=16)
plt.grid()


plt.show()
```

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]


for n in range(100):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        print("Loop ran this many times", n)
        break

print('Local min occure at:', new_x)
print("Slope or df(x) value at this point is:",df(new_x))
print("f(x) value or cost at this point is:",f(new_x))
```

```python
#superimpose gradient decent data

plt.figure(figsize=[20,5])

# 1st chart cost function

plt.subplot(1,3,1)
plt.title("cost function")
plt.plot(x_1,f(x_1),color="blue",linewidth=3,alpha=0.6)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel("x",fontsize=16)
plt.ylabel("f(x)",fontsize=16)

values = np.array(x_list)

plt.scatter(x_list,f(values),color="red",s=100,alpha=0.6)

#2nd chart for derative

plt.subplot(1,3,2)
plt.title("Slope of cost function")
plt.xlim([-2,3])
plt.ylim(-3,6)
plt.plot(x_1,df(x_1),color="skyblue",linewidth=3,alpha = 0.6)
plt.xlabel("x",fontsize=16)
plt.ylabel("d(x)",fontsize=16)
plt.grid()
plt.scatter(x_list,slope_list,color="red",s=100,alpha=0.5)


plt.show()

#3nd chart for derative

plt.subplot(1,3,3)
plt.title("Gradient Decent")
plt.xlim([-0.55,-0.2])
plt.ylim(-0.3,0.8)
plt.plot(x_1,df(x_1),color="skyblue",linewidth=3,alpha = 0.6)

plt.grid()
plt.scatter(x_list,slope_list,color="red",s=100,alpha=0.5)


plt.show()
```

```python
# g(x) = x^4 - 4x^2 + 5

x_2 = np.linspace(-2,2,1000)

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=[15,5])

# 1st chart cost function

plt.subplot(1,2,1)
plt.title("cost function")
plt.plot(x_2,g(x_2),color="blue",linewidth=3)
plt.xlim([-2,2])
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize=16)
plt.ylabel("g(x)",fontsize=16)

#2nd chart for derative

plt.subplot(1,2,2)
plt.title("slope of cost function")
plt.xlim([-2,2])
plt.ylim(-6,8)
plt.plot(x_2,dg(x_2),color="skyblue",linewidth=3)
plt.xlabel("x",fontsize=16)
plt.ylabel("dg(x)",fontsize=16)
plt.grid()


plt.show()
```

```python
new_x = -0.8
previous_x = 0
step_multiplier = 0.2
precision = 0.001

x_list = [new_x]
slope_list = [dg(new_x)]


for n in range(5000):
    previous_x = new_x
    gradient = dg(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    x_list.append(new_x)
    slope_list.append(dg(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        break
```

```python
#superimpose gradient decent data

plt.figure(figsize=[20,5])

# 1st chart cost function

plt.subplot(1,3,1)
plt.title("cost function")
plt.plot(x_2,g(x_2),color="blue",linewidth=3,alpha=0.6)
plt.xlim([-2,2])
plt.ylim(0.5,5.5)
plt.xlabel("x",fontsize=16)
plt.ylabel("g(x)",fontsize=16)

values = np.array(x_list)

plt.scatter(x_list,g(values),color="red",s=100,alpha=0.6)

#2nd chart for derative

plt.subplot(1,3,2)
plt.title("Slope of cost function")
plt.xlim([-2,2])
plt.ylim(-6,8)
plt.plot(x_2,dg(x_2),color="skyblue",linewidth=3,alpha = 0.6)
plt.xlabel("x",fontsize=16)
plt.ylabel("dg(x)",fontsize=16)
plt.grid()
plt.scatter(x_list,slope_list,color="red",s=100,alpha=0.5)


plt.show()

#3nd chart for derative

plt.subplot(1,3,3)
plt.title("Gradient Decent")
plt.xlim([-0.55,-0.2])
plt.ylim(-0.3,0.8)
plt.plot(x_2,dg(x_2),color="skyblue",linewidth=3,alpha = 0.6)

plt.grid()
plt.scatter(x_list,slope_list,color="red",s=100,alpha=0.5)


plt.show()
```

```python
new_x = 0.01
previous_x = 0
step_multiplier = 0.2
precision = 0.001

x_list = [new_x]
slope_list = [dh(new_x)]


for n in range(500):
    previous_x = new_x
    gradient = dh(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    x_list.append(new_x)
    slope_list.append(dh(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        break
```
